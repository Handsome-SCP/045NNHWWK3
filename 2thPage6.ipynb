```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
#Sample data: Movie genre ratings (Action, Comedy, Drama, Sci-Fi)
X = np.array([[8, 2, 4, 7],
              [5, 7, 3, 1],
              [2, 8, 7, 2],
              [9, 1, 3, 8],
              [3, 6, 9, 2],
              [7, 3, 2, 9]])
movie_names = ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E', 'Movie F']
```

```python
#Perform thierarchical clustering
linkage_matrix = linkage(X, method='ward')
```

```python
#Plot dendragram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix, labels=movie_names)
plt.title('Movie Clustering Dendrogram')
plt.xlabel('Movies')
plt.ylabel('Distance')
plt.show()
```

```python
#Create cluster
n_clusters = 3
clustering = AgglomerativeClustering(n_clusters=n_clusters)
cluster_labels = clustering.fit_predict(X)
```

```python
#Print results
for i in range(n_clusters):
    cluster_movies = [movie_names[j] for j in range(len(movie_names)) if cluster_labels[j] == i]
    print(f"Cluster {i+1}: {', '.join(cluster_movies)}")
```
